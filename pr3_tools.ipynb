# **Row denormaliser**
используется для приведения строк данных в денормализованную форму, когда несколько строк объединяются в одну строку. Обычно он используется для преобразования данных из нормализованной структуры (несколько строк со связанными значениями) в денормализованную структуру (одна строка с несколькими столбцами, представляющими разные связанные значения)
